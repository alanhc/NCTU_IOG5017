```python
freshman = df[df.等級別 == "B 學士"].一年級男生
freshman
```

```python
df['一年級男女比'] = df.一年級男生/df.一年級女生
```

```python
df[["學校名稱", "一年級男女比", "一年級男生","一年級女生"]].sort_values(by="一年級男女比")
```

```python
big_school = df.一年級男生 + df.一年級女生 > 1300
C = df.等級別 == "B 學士"
df[C & big_school][["學校名稱", "一年級男女比", "一年級男生","一年級女生"]].sort_values(by="一年級男女比")
```

```python
result = df[C & big_school][["學校名稱", "一年級男女比"]].sort_values(by="一年級男女比")
```

```python
result.plot.bar()
```
